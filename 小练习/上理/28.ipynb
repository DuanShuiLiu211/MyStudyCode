```python
import matplotlib.pyplot as plot
import numpy as np
```

```python
x = np.arange(-5, 5, 0.001)


def func1(x):
    return 1 - ((2 * (1 / (1 + np.exp(-x)))) / (1 + (1 / (1 + np.exp(-x)))))


def func2(x):
    return (
        (-2)
        / (1 + (1 / (1 + np.exp(-x)))) ** 2
        * ((np.exp(-x)) / (1 + np.exp(-x)) ** 2)
    )


def func3(x):
    return 1 / (1 + np.exp(-x))


plot.figure(figsize=(10, 5))
plot.plot(x, func1(x))
plot.plot(x, func2(x))
plot.plot(x, func3(x))
plot.grid(True, linestyle="--", alpha=0.5)
plot.show()
```
