```python
# 0. 生成器协程实现生产者消费者模型
def consumer():
    r = ""
    while True:
        n = yield r  # yield的返回值为r接受值赋予n
        if not n:
            return
        print("[CONSUMER] Consuming %s..." % n)
        r = "200 OK"


def produce(c):
    c.send(None)
    n = 0
    while n < 5:
        n = n + 1
        print("[PRODUCER] Producing %s..." % n)
        r = c.send(n)  # 通知已经完成
        print("[PRODUCER] Consumer return: %s" % r)
    c.close()


if __name__ == "__main__":
    c = consumer()
    produce(c)
```

```python
# 1. 生成器测试
def my_generator():
    print("Start")
    yield 1
    print("Middle")
    yield 2
    print("End")
    yield


gen = my_generator()
result1 = gen.send(None)  # 启动生成器，执行到第一个 yield，输出 "Start"
result2 = gen.send(None)  # 执行到第二个 yield，输出 "Middle"
gen.send(None)  # 输出 "End"，执行最后一个 yield，结束程序

print(result1)  # 输出 1
print(result2)  # 输出 2


class Fib(object):
    def __init__(self, num):
        super(Fib, self).__init__()
        self.max = num
        self.x = 0
        self.y = 1

    def __iter__(self):
        return self

    def __next__(self):
        fib = self.x
        if fib > self.max:
            raise StopIteration
        self.x, self.y = self.y, self.x + self.y  # f(n), f(n+1) = f(n+1), f(n) + f(n+1)
        return fib


def main():
    fib = Fib(5)
    print(fib)
    for ans in fib:
        print(ans)


if __name__ == "__main__":
    main()
    fib = Fib(5)
    print(fib)
    print(next(fib))
    print(next(fib))
    print(next(fib))
    print(next(fib))
    print(next(fib))
```
