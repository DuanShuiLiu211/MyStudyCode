```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as animation
```

```python
# Defining x axis and y = f(x)
x = np.linspace(0, 1, 500)

# y = signal.square(2 * np.pi * 5 * x)
# y = signal.sawtooth(2 * np.pi * 5 * x)
# y = np.sin(2*np.pi*x) + np.random.normal(scale=0.3, size=len(x))
y = np.cos(4 * np.pi * x) + np.random.normal(scale=0.15, size=len(x))

n = len(y)
plt.plot(x, y)
```

```python
# Plotting the generated figure with n = 0 to 300 coefficients and generating the animated GIF
n = 1000
x = np.linspace(0, 10, n)
y = np.sin(x)

fig, ax = plt.subplots(figsize=(10, 10))


def animate(i):
    ax.clear()
    Y = np.fft.fft(y) / n * 1000
    np.put(Y, range(i + 1, n), 0.0)
    ifft = np.abs(np.fft.ifft(Y))
    ax.plot(x, ifft, color="r", linewidth=1.5)
    ax.plot(x, y, color="k")
    ax.legend([f"n = {i}"], loc=3)
    ax.grid(True, axis="both")


ani = animation.FuncAnimation(fig, animate, frames=300, interval=3)
ani.save("fourier.gif", writer="pillow")
```
