```python
x_data = [1.0, 2.0, 3.0]
y_data = [2.0, 4.0, 6.0]

w = torch.Tensor([1.0])  # åˆå§‹æƒå€¼
w.requires_grad = True  # è®¾ç½®ä¸ºéœ€è¦è®¡ç®—æ¢¯åº¦


def forward(x):
    y_pred = x * w
    return y_pred


def loss(y_pred, y):
    l = (y_pred - y) ** 2
    return l


def optim(w):
    w.data = w.data - 0.01 * w.grad.data


print('Predict (befortraining)', 4, forward(4).item())

for epoch in range(100):
    print('Epoch:', epoch)
    for x, y in zip(x_data, y_data):
        if w.grad is not None:
            w.grad.data.zero_()  # ç­‰ä»·äºŽ optimizer.zero_grad()
        y_pred = forward(x)
        l = loss(y_pred, y)
        l.backward()
        print(f'x is {x}, y is {y}ï¼Œl is {l.item()}')
        print(f'w is {w.item()}, w grad is {w.grad.item()}')
        optim(w)  # ç­‰ä»·äºŽ optimizer.step() çš„ä½œç”¨

print('Predict(after training)', 4, forward(4).item())
```
